```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_excel('./results_shd_soa.xls', sheet_name='D-SNN')
df_rsnn = pd.read_excel('./results_shd_soa.xls', sheet_name='R-SNN')
style = "whitegrid"
figsize = (7,5)
```

```python
df_rsnn
```

```python
df
```

```python
df_np = pd.concat([df[df['pruning']==50],  df[df['pruning']==25] ]) 
```

```python
df_np
```

```python
df_p = df[(df['pruning']!=50) & (df['pruning']!=25)]
```

```python
df_p
```

```python
df_lif = df_rsnn[df_rsnn['type']=='LIF']
df_alif = df_rsnn[df_rsnn['type']=='ALIF']
```

#### Pruning vs No Pruning
(run twice to fix the display of first plot, seaborn bug?)

```python
palette = sns.color_palette("crest",5)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)

sns.set_style(style)
g=sns.scatterplot(data = df_alif, x="params", y="acc", color='orange', label='rsnn (ALIF)', legend=False)
sns.scatterplot(data = df_lif, x="params", y="acc", color='red', label='rsnn (LIF)', legend=False)
sns.scatterplot(data = df_p, x="params", y="acc",hue='pruning', palette= palette, legend=False)
sns.scatterplot(data = df_np, x="params", y="acc", color='k', label='no pruning', legend=False)
plt.xscale("log")
plt.xlabel('number of synaptic parameters (millions)')
#xlabels = [x for x in g.get_xticks()/1000000]
#g.set_xticklabels(xlabels)

#plt.xlim(0.1,600000)
plt.xlim(0.0,2e6)

plt.subplot(1,2,2)
sns.scatterplot(data = df_alif, x="spk_time", y="acc", color='orange', label='rsnn (ALIF)')
sns.scatterplot(data = df_lif, x="spk_time", y="acc", color='red', label='rsnn (LIF)')
sns.scatterplot(data = df_p, x="spk_time", y="acc",hue='pruning', palette= palette)

sns.scatterplot(data = df_np, x="spk_time", y="acc", color='k', label='no pruning')
plt.xlabel('spikes per timestep')
plt.savefig('shd_results.pdf', dpi=200)
plt.xlim(3,26)
plt.savefig('shd_pruning_acc.pdf', dpi=200)
```
